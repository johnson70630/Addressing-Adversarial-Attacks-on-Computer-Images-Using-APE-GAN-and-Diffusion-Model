```python
# Define device
import torch
if torch.cuda.is_available():
  device = torch.device('cuda')
else:
  device = torch.device('cpu')
print('Device: ', device)
```

```python
train_data = torch.load("data.tar")
x_adv = train_data["adv"]
x_adv = x_adv.to(device)
```

```python
import os

check_path = "./AdvImages"
os.makedirs(check_path, exist_ok=True)
```

```python
x_adv.shape
```

```python
from PIL import Image

for i, image_data in enumerate(x_adv):
  image = Image.fromarray(image_data[0].cpu().numpy() * 255).convert('L')  # 將圖像從Tensor轉換為PIL格式
  filename = os.path.join(check_path, f"adv_image_{i + 1}.png")
  image.save(filename)

print(f"Save {len(x_adv)} Pictures in {check_path} OK")
```
